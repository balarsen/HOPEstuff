```python
import numpy as np
import pymc as mc
import matplotlib.pyplot as plt
```

```python
dat = {}
for counts in np.arange(1, 1000):
    print("ITER", counts)
    N = mc.Poisson('N', float(counts))
    model = mc.MCMC((N,))
    model.sample(20000)
    dat[counts] = np.percentile(N.trace(), (5, 10, 20, 25,50, 75,80,90,95))
```

```python
for d in dat:
    print(d, dat[d])
#     plt.errorbar(d, d, dat[d].T)
```

```python
import spacepy.datamodel as dm
data = dm.SpaceData()
data['percentiles'] = dm.dmarray((5, 10, 20, 25,50, 75,80,90,95))
data['value'] = dm.dmarray(np.zeros(len(dat)+1))
data['error'] = dm.dmarray(np.zeros((len(data['value']), 
                                     len(data['percentiles']))))
for d in dat:
    data['value'][d] = d
    data['error'][d] = dat[d]
```

```python
data.tree(verbose=1)
data['error'].attrs['DEPEND_0'] = 'value'
data['error'].attrs['DEPEND_1'] = 'percentiles'
```

```python
dm.toHDF5('Poisson.h5', data, compression='gzip')
```
