```python
# counts are 100 in all directions
polar_edges = np.linspace(-3,3,21)
azi_edges = np.linspace(-4,4, 23)
counts2d, weights2d = pixel_matching_rel04.resp_2d(100, polar_edges, azi_edges, to_arr=False)
ans = pixel_matching_rel04.resp_2d_to_arr(polar_edges, azi_edges, counts2d, weights2d)
# f, (ax1, ax2, ax3) = plt.subplots(1,3, sharex=False, sharey=False)
fig = plt.figure(figsize=(15,5))
ax = fig.add_subplot(131)
pc = ax.pcolormesh(polar_edges, azi_edges, counts2d)
ax.set_aspect('equal', 'datalim')
fig.colorbar(pc)
ax.set_xlabel('Polar')
ax.set_ylabel('Azi')
ax.set_title('Counts')

ax = fig.add_subplot(132, sharey=ax)
pc = ax.pcolormesh(polar_edges, azi_edges, weights2d)
ax.set_aspect('equal', 'datalim')
ax.set_xlabel('Polar')
plt.colorbar(pc)
ax.set_title('Weights')

ax = fig.add_subplot(133, sharey=ax)
pc = ax.pcolormesh(polar_edges, azi_edges, counts2d/weights2d)
ax.set_aspect('equal', 'datalim')
ax.set_xlabel('Polar')
plt.colorbar(pc)
ax.set_title('Normalized Counts')


```
